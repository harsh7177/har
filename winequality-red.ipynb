```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
wine=pd.read_csv("winequality-red.csv")
wine.head()
```

```python
wine.isnull().sum()
```

```python
wine.describe(include="all")
```

```python
plt.figure(figsize=(8,8))
sns.heatmap(wine.corr())
```

```python
wine.alcohol.value_counts().plot(kind='bar')
```

```python
sns.countplot(x='quality', data=wine)
```

```python
correlation = wine.corr()
plt.figure(figsize=(14, 8))
sns.heatmap(correlation, annot=True, linewidths=0, vmin=-1, cmap="RdBu_r")
```

```python
sns.lmplot(x = "alcohol", y = "residual_sugar",  data = wine)
```

```python
sns.scatterplot(data=wine,x="sulphates",y='alcohol',hue="quality")
```

```python
sns.scatterplot(data=wine,x="sulphates",y='alcohol',hue="pH",size="quality")
```

```python
sns.relplot(data=wine,x="sulphates",y='alcohol',hue="pH",col="quality")
```
